## Data cleaning
Dataset will be saved 2 times with different preprocessing:
    1. Data with combined categorical data for `funder` and `installer`
    2. The same as 1, but with removed outliers, and dropped all missing values

`amount_tsh` - 41639 of 59400 values missing: better to remove whole column

```python
data.drop('amount_tsh', axis=1, inplace=True)
```

#### Split dataset first to avoid data leakage

```python
train, test = train_test_split(data,random_state=42)
```

#### 1. Features with combined categorical data
Decrease unique values of selected columns to 100 (set value 'Other' for less frequent columns)

```python
train_funder_col_names, assigned_val, train.funder = hlp.reduce_cat_num(train.funder,'Other')
test.funder = hlp.apply_reduced_cats(test.funder,train_funder_col_names,assigned_val)

train_installer_col_names, assigned_val, train.installer = hlp.reduce_cat_num(train.installer,'Other')
test.installer = hlp.apply_reduced_cats(test.installer,train_installer_col_names,assigned_val)
```

#### Save

```python
with open('data/preprocessed/train_comb_cats.pickle', 'wb') as file:
    pickle.dump(train, file)
    
with open('data/preprocessed/test_comb_cats.pickle', 'wb') as file:
    pickle.dump(test, file)
```

#### 2. Features with combined categorical data, removed outliers, and dropped all missing values

```python
# train dropped missing values
train_dmv = train.copy()
train_dmv = train_dmv[train_dmv['gps_height'] != 0]
train_dmv = train_dmv[train_dmv['longitude'] != 0]
train_dmv = train_dmv[train_dmv['population'] != 0]
train_dmv = train_dmv[train_dmv['construction_year'] != 0]

hlp.print_reduce_perc(train,train_dmv)
```

```python
# Check for -2.000000e-08 value
train_dmv.latitude.value_counts()
```

#### Remove outliers

```python
possible_outliers_cols = ['gps_height','longitude','latitude','population']
train_dmv_cleaned = train_dmv[(np.abs(stats.zscore(train_dmv[possible_outliers_cols])) < 3).all(axis=1)]

hlp.print_reduce_perc(train_dmv,train_dmv_cleaned)
```

#### Save

```python
with open('data/preprocessed/train_comb_cats_cleaned.pickle', 'wb') as file:
    pickle.dump(train_dmv_cleaned, file)
```
